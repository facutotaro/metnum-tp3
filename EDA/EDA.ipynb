## Correlaciones

```python
def pretty_qqplot(df, feature_a, feature_b):
    
    percs = np.linspace(1,99,99)
    qn_a = np.nanpercentile(df[feature_a], percs)
    qn_b = np.nanpercentile(df[feature_b], percs)

    plt.plot(qn_a,qn_b, ls="", marker="o")

    plt.plot([qn_a[0],qn_a[-1]],[qn_b[0],qn_b[-1]])

    plt.title(feature_a + " vs " + feature_b, fontsize = 20)
    plt.xlabel(feature_a, fontsize = 15)
    plt.ylabel(feature_b, fontsize = 15)
```

```python
df_corr = df.corr()
df_corr
```

## Pares de alta correlacion

```python
columns = df.columns[3:-1]

pares_corr = []

for cc in columns:
    
    for cf in columns:
        
        if abs(df_corr[cc][cf]) > 0.7 and cc != cf and not((cf,cc) in pares_corr):
            
            pares_corr.append((cc,cf))
            
            print(cc + " - " + cf + " : " + str(df_corr[cc][cf]))
```

```python
plt.figure(figsize = (30, 55))

x = 7
y = 2

for j in range(y):
    
    for i in range(x):
    
        plt.subplot(x, y, i + 1 + (x * j))
        pretty_qqplot(df, pares_corr[(j*x) + i][0], pares_corr[(j*x) + i][1])
        
plt.tight_layout()
plt.show()
```

## Datos con poca correlacion

```python
columns = df.columns[3:-1]

no_corr = []

for cc in columns:
    
    res = True
    maximo = 0
    cf_max = ""
    
    for cf in columns: 
        
        if abs(df_corr[cc][cf]) > 0.7 and cc != cf: res = False
        if abs(df_corr[cc][cf]) > maximo and cc != cf:
            
            maximo = df_corr[cc][cf]
            cf_max = cf
            
    if res: 
        
        no_corr.append(cc)
        
        print(cc + " : " + cf_max + " " + str(maximo))
```

## Caracteristicas por expectativa de vida

```python
# reemplazamos valores nulos por promedio de la feature
values = {}

cols = list(df.columns[3:-1])

print(cols)
for c in cols:
    
    values[c] = df[c].mean()

df = df.fillna(value = values)
```

```python
grupo_b = df[df["Life expectancy"] < df["Life expectancy"].quantile(0.1)]
```

```python
grupo_a = df[df["Life expectancy"] > df["Life expectancy"].quantile(0.9)]
```

```python
import matplotlib
def pretty_barplot(df_a, df_b, name_a, name_b, feature):

    plt.bar([name_a,name_b],[grupo_a[feature].mean(), grupo_b[feature].mean()])
    plt.title(feature, fontsize = 30)
    plt.ylabel("cantidad de paises", fontsize = 15)
    
def pretty_double_boxplot(df_a, df_b, name_a, name_b, feature):
    
    plt.boxplot([grupo_a[feature], grupo_b[feature]])
    plt.title(feature, fontsize = 15)
    plt.xticks([1,2],[name_a,name_b])
```

```python
features = df.columns[3:-1]

plt.figure(figsize = (30, 55))

font = {'family' : 'DejaVu Sans',
        'weight' : 'bold',
        'size'   : 15}

matplotlib.rc('font', **font)

x = 9
y = 2

for j in range(y):
    
    for i in range(x):
    
        plt.subplot(x, y, i + 1 + (x * j))
        pretty_double_boxplot(grupo_a, grupo_b,"grupo_a", "grupo_b", features[(j*9) + i])
        
plt.tight_layout()
plt.show()
```

```python
features = df.columns[3:-1]


for f in features:

    plt.figure(figsize = (5, 3))
    
    pretty_double_boxplot(grupo_a, grupo_b, "grupo_a", "grupo_b", f)
    
    if f == "HIV/AIDS": plt.savefig("images/HIV_AIDS" + "_grous")
    else: plt.savefig("images/" + f + "_groups", bbox_inches = "tight")
        
    plt.clf()
```

```python
def pretty_scatter(df_ped, df_ping, feature):
    x = df_ped[df_ped["BMI"] > 30] #Filtrado leverage
    x = x[x["Hepatitis B"] > 60][feature]
    
    y = df_ped[df_ped["BMI"] > 30]
    y = y[y["Hepatitis B"] > 60]["Life expectancy"]

    x2 = df_ping[feature]
    y2 =  df_ping["Life expectancy"]
    
    x3 = df[feature]
    y3 = df["Life expectancy"]
    
    plt.plot(x, y, 'o')

    m, b = np.polyfit(x, y, 1)

    plt.plot(x2, y2,'o')
    
    m3, b3 = np.polyfit(x3, y3, 1)
    
    m1, b1 = np.polyfit(x2, y2, 1)
    
    develo = "develo"
    
    plt.title(feature)
    plt.plot(x, m*x + b, label = develo + "ped")
    plt.plot(x2, m1*x2 + b1, label = develo + "ping")
    plt.plot(x3, m3*x3 + b3, label = "all")
    plt.legend()
    
    plt.xlabel(feature)
    plt.ylabel("life expectancy")
```

```python
features = df.columns[3:-1]

plt.figure(figsize = (20, 55))

font = {'family' : 'DejaVu Sans',
        'weight' : 'bold',
        'size'   : 15}

matplotlib.rc('font', **font)

x = 9
y = 2

for j in range(y):
    
    for i in range(x):
    
        plt.subplot(x, y, i + 1 + (x * j))
        pretty_scatter(df_developed, df_developing, features[(j*9) + i])
        
plt.tight_layout()
plt.show()
```

# Alcohol es Confounding

```python
plt.figure(figsize = (20,10))

plt.subplot(2,2,1)
plt.title("Alcohol vs GDP")
sns.scatterplot(data = df, x = "GDP", y = "Alcohol", hue = "Status")

m, b = np.polyfit(df["GDP"], df["Alcohol"], 1)
x = np.linspace(0,60000,1000)
plt.plot(x, m*x + b, c = "green")

plt.legend(loc = "lower right")

plt.subplot(2,2,2)
plt.title("life expectancy vs GDP")
sns.scatterplot(data = df, x = "GDP", y = "Life expectancy", hue = "Status")
plt.legend(loc = "lower right")

m, b = np.polyfit(df["GDP"], df["Life expectancy"], 1)
x = np.linspace(0,60000,1000)
plt.plot(x, m*x + b, c = "green")

plt.subplot(2,2,3)
plt.title("life expectancy vs Alcohol")
sns.scatterplot(data = df, x = "Alcohol", y = "Life expectancy", hue = "Status")
plt.legend(loc = "lower right")

m, b = np.polyfit(df["Alcohol"], df["Life expectancy"], 1)
x = np.linspace(0,14,1000)
plt.plot(x, m*x + b, c = "green")

plt.tight_layout()
plt.show()
```
